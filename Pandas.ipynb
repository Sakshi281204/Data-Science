```python
import pandas as pd                                                       # Import pandas library

# Area Series
area = pd.Series(
    [307713, 243286, 38863, 196024, 50362,33788],                                # List of state areas in sq km
    index=['Maharashtra', 'Uttar Pradesh', 'Kerala', 'Gujarat', 'Punjab','Goa']  # State names as index
)

# Population Series
population = pd.Series(
    [123, 231, 35, 70, 30],                                                # Population in millions
    index=['Maharashtra', 'Uttar Pradesh', 'Kerala', 'Gujarat', 'Punjab']
)

# Create DataFrame combining area and population7m
data = pd.DataFrame({'Area_sq_km': area,'Population_million': population})

# Compute density (people per sq km)
data['Density_per_sq_km'] = data['Population_million'] * 1_000_000 / data['Area_sq_km']
print(data)
mean_density = data['Density_per_sq_km'].std()

# Step 2: Fill Goa's density
data.loc['Goa', 'Density_per_sq_km'] = mean_density

# Step 3: Calculate Goa's population: area × density → convert to millions
data.loc['Goa', 'Population_million'] = (data.loc['Goa', 'Area_sq_km'] * data.loc['Goa', 'Density_per_sq_km']) / 1_000_000
print(data)
```

```python
import seaborn as sns
import pandas as pd

# Load dataset and drop missing age
Titanic = sns.load_dataset('titanic')
Titanic = Titanic.dropna(subset=['age'])

# Add a column for count=1 to use in groupby
Titanic['count'] = 1

# Define helper function for survived and not survived data
def make_subtable(survived_flag, agg_funcs):
    df = Titanic[Titanic['survived'] == survived_flag]
    return df.groupby(['sex', 'class']).agg(agg_funcs).unstack(level='class')

# Survived part: we need 'age': 'sum' and 'count': 'sum'
survived_table = make_subtable(1, {'age': 'sum', 'count': 'sum'})

# Not Survived part: we need 'age': 'mean' and 'count': 'sum'
not_survived_table = make_subtable(0, {'age': 'mean', 'count': 'sum'})

# Rename columns for clarity
survived_table.columns = pd.MultiIndex.from_product([['survived'], survived_table.columns.levels[1], survived_table.columns.levels[0]])
not_survived_table.columns = pd.MultiIndex.from_product([['not survived'], not_survived_table.columns.levels[1], not_survived_table.columns.levels[0]])

# Combine both tables
final_table = pd.concat([survived_table, not_survived_table], axis=1)

# Sort columns to match visual order
final_table = final_table[
    pd.MultiIndex.from_product([
        ['survived', 'not survived'],
        ['First', 'Second', 'Third'],
        ['age', 'count'] if 'survived' in final_table.columns.get_level_values(0) else ['mean', 'count']
    ])
]

# Rename classes to match display format
final_table.columns = final_table.columns.set_levels(
    ['first class', 'second class', 'third class'],
    level=1
)

# Display result
print(final_table)
```
