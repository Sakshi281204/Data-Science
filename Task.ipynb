# **PROBLEM STATEMENT**
Study and Compare the Behavior of Big Tech Company Stocks Over Time
Analyze stock price data from companies like Apple, Google, and Amazon to understand patterns, trends, and volatility. Focus on how these stocks behave over months or years, and figure out which one is more stable or more profitable during different time periods.


```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler

# Step 1: Function to fetch and preprocess stock data
def fetch_stock_data(tickers, start_date='2010-01-01', end_date='2023-01-01'):
    data = {}
    for ticker in tickers:
        stock_data = yf.download(ticker, start=start_date, end=end_date)
        stock_data = stock_data.fillna(method='ffill')
        stock_data['Daily_Return'] = stock_data['Close'].pct_change()
        monthly_data = stock_data.resample('M').mean()
        monthly_data['Monthly_Return'] = monthly_data['Close'].pct_change()
        data[ticker] = monthly_data
    return data

tickers = ['AAPL', 'GOOGL', 'AMZN']
stock_data = fetch_stock_data(tickers)

def plot_stock_trends(stock_data):
    plt.figure(figsize=(12, 6))
    for ticker, data in stock_data.items():
        plt.plot(data.index, data['Close'], label=ticker)
    plt.title('Stock Price Trends Over Time')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (Monthly Average)')
    plt.legend()
    plt.show()

plot_stock_trends(stock_data)

def analyze_stock_behavior(stock_data):
    results = {}
    for ticker, data in stock_data.items():
        average_daily_return = data['Daily_Return'].mean() * 252
        volatility = data['Daily_Return'].std() * np.sqrt(252)
        cumulative_return = (1 + data['Daily_Return']).cumprod().iloc[-1] - 1
        results[ticker] = {
            'Average Annual Return': average_daily_return,
            'Annual Volatility': volatility,
            'Cumulative Return': cumulative_return
        }
        plt.figure(figsize=(8, 4))
        plt.hist(data['Daily_Return'].dropna(), bins=50, label=ticker, alpha=0.5)
        plt.title(f'Daily Returns Distribution for {ticker}')
        plt.xlabel('Daily Return')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()
    return results

behavior_results = analyze_stock_behavior(stock_data)
print("Overall Stock Behavior Summary:")
for ticker, metrics in behavior_results.items():
    print(f"{ticker}: Average Annual Return = {metrics['Average Annual Return']:.2%}, "
          f"Annual Volatility = {metrics['Annual Volatility']:.2%}, "
          f"Cumulative Return = {metrics['Cumulative Return']:.2%}")

def compare_over_periods(stock_data, periods):
    comparisons = {}
    for period in periods:
        start, end = period
        period_data = {ticker: data.loc[start:end] for ticker, data in stock_data.items()}
        period_results = analyze_stock_behavior(period_data)
        comparisons[period] = period_results
        print(f"\nComparison for period {start} to {end}:")
        for ticker, metrics in period_results.items():
            print(f"{ticker}: Avg Annual Return = {metrics['Average Annual Return']:.2%}, "
                  f"Volatility = {metrics['Annual Volatility']:.2%}")
        most_profitable = max(period_results, key=lambda x: period_results[x]['Average Annual Return'])
        most_stable = min(period_results, key=lambda x: period_results[x]['Annual Volatility'])
        print(f"Most Profitable: {most_profitable}, Most Stable: {most_stable}")
    return comparisons

periods = [('2010-01-01', '2015-12-31'), ('2016-01-01', '2020-12-31'), ('2021-01-01', '2023-01-01')]
comparison_results = compare_over_periods(stock_data, periods)

def build_and_evaluate_ml_model(stock_data):
    models = {}
    mae_dict = {}
    for ticker, data in stock_data.items():
        data = data.dropna()
        data['Lagged_Return_1'] = data['Daily_Return'].shift(1)
        data['Lagged_Return_5'] = data['Daily_Return'].shift(5)
        features = data[['Lagged_Return_1', 'Lagged_Return_5']].dropna()
        target = data['Daily_Return'].shift(-1).dropna()

        if len(features) > 0 and len(target) > 0:
            X = features
            y = target[:len(X)]
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
            scaler = StandardScaler()
            X_train_scaled = scaler.fit_transform(X_train)
            X_test_scaled = scaler.transform(X_test)
            model = LinearRegression()
            model.fit(X_train_scaled, y_train)
            models[ticker] = model
            y_pred = model.predict(X_test_scaled)
            mae = mean_absolute_error(y_test, y_pred)
            print(f"{ticker}: Mean Absolute Error (MAE) = {mae:.4f}")
            mae_dict[ticker] = mae
        else:
            print(f"Not enough data for {ticker}")
            mae_dict[ticker] = None
    return models, mae_dict

ml_models, mae_dict = build_and_evaluate_ml_model(stock_data)

print("\nFinal Insights:")
for period, results in comparison_results.items():
    print(f"For period {period[0]} to {period[1]}:")
    for ticker in tickers:
        metrics = results[ticker]
        print(f"  {ticker}: Profitability (Avg Return) = {metrics['Average Annual Return']:.2%}, "
              f"Stability (Volatility) = {metrics['Annual Volatility']:.2%}")

    valid_mae_tickers = {ticker: mae for ticker, mae in mae_dict.items() if mae is not None}
    if valid_mae_tickers:
        most_stable_from_ml = min(valid_mae_tickers, key=valid_mae_tickers.get)
        print(f"  Most Stable based on ML (lowest prediction error): {most_stable_from_ml}")
    else:
        print("  No valid ML results for stability comparison.")
```

```python
# --- User Input for Prediction ---
print("\n--- STOCK RETURN PREDICTION ---")
user_ticker = input("Enter the stock symbol to predict (AAPL / GOOGL / AMZN): ").upper()

if user_ticker in ml_models:
    data = stock_data[user_ticker].dropna()

    # Ensure lagged features exist
    data['Lagged_Return_1'] = data['Daily_Return'].shift(1)
    data['Lagged_Return_5'] = data['Daily_Return'].shift(5)
    data = data.dropna()

    # Get the most recent values for prediction
    latest_features = data[['Lagged_Return_1', 'Lagged_Return_5']].iloc[-1].values.reshape(1, -1)

    # Scale using StandardScaler (fit on full data for simplicity)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(data[['Lagged_Return_1', 'Lagged_Return_5']])
    latest_scaled = scaler.transform(latest_features)

    # Predict next day's return
    model = ml_models[user_ticker]
    predicted_return = model.predict(latest_scaled)[0]

    # Convert predicted return to % format
    print(f"\nPredicted next day's return for {user_ticker}: {predicted_return * 100:.3f}%")

    # Optional: visualize the recent return trend
    plt.figure(figsize=(8, 4))
    plt.plot(data.index[-30:], data['Daily_Return'].iloc[-30:], label='Past Daily Returns', color='blue')
    plt.axhline(predicted_return, color='red', linestyle='--', label='Predicted Next Return')
    plt.title(f"Recent 30-Day Returns and Prediction for {user_ticker}")
    plt.xlabel("Date")
    plt.ylabel("Daily Return")
    plt.legend()
    plt.show()

else:
    print("Invalid ticker. Please enter one of: AAPL, GOOGL, or AMZN.")
```
